# 批处理

```python
import os 
import pandas as pd
keywords_path = "./dataset/keywords/"
for _ , _ ,files in os.walk(keywords_path):
    writer = pd.ExcelWriter("final.xlsx")
    for file in files:
        keyname = file.split("-")[1].lower()
        path = keywords_path+file
        df = pd.read_excel(path)
        df = df[df["关键词"].str.contains(keyname)].sort_values(by="月搜索量",ascending=False).reset_index(drop=True)
        df = df[df["月搜索量"]>10000]
        df = df[["关键词","月搜索量","所属类目"]]
        if keyname=="shark":
            df = df[~df["关键词"].str.contains("vacuum")]
            df = df[~df["关键词"].str.contains("keto")]
            df = df[~df["关键词"].str.contains("mop")]
            df = df[~df["关键词"].str.contains("hv")]
            df = df[~df["关键词"].str.contains("weight")]
        df.to_excel(writer,sheet_name=keyname)
    writer.close()
```
