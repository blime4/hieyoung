```python
import os
import random
from PIL import Image, ImageTk    
import tkinter as tk    


def resize(w_box,h_box,pil_image):
    w,h = pil_image.size
    f1 = 1.0*w_box/w
    f2 = 1.0*h_box/h
    factor = min([f1,f2])
    width = int(w*factor)
    height = int(h*factor)
    return pil_image.resize((width,height),Image.ANTIALIAS)

root = tk.Tk() 
w_box = 300    #期望图像显示的大小（窗口大小）
h_box = 200
pil_image = Image.new('RGB', (300, 200), (255,255,255))
tk_image = ImageTk.PhotoImage(pil_image)
label = Label(root,image=tk_image,width=w_box,height=h_box)
label.pack(padx=5,pady=5)
def change():
    rootdir = "./png_get/"
    file_names = os.listdir(rootdir)
    x = random.randint(0, len(file_names)-1)
    pil_image = Image.open(rootdir+file_names[x])
    pil_image_resized = resize(w_box,h_box,pil_image)
    tk_image = ImageTk.PhotoImage(pil_image_resized)
    label.configure(image=tk_image)
    label.image = tk_image
button = Button(root,text="change",command=change).pack()
root.mainloop()
```
