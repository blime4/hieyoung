```python
comments_token ={}
for comment in comments:
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(comment)
    tokens = [i for i in tokens if(i.lower() not in stopwords.words('english'))]
    for token in tokens:
        if token not in comments_token:
            comments_token[token] = 1
        else:
            comments_token[token] += 1
df_comments_token = pd.DataFrame.from_dict(comments_token,orient="index",columns=["频率"]).sort_values(by="频率",ascending=False)
df_comments_token = df_comments_token[df_comments_token["频率"]>5]
df_comments_token
```

```python
from nltk.util import ngrams 

bigrams_titles = {}
titles_token ={}
for title in titles:
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(title)
    tokens = [i for i in tokens if(i.lower() not in stopwords.words('english'))]
    bigrams_t = ngrams(tokens, 2) 
    for b in bigrams_t:
        if b not in bigrams_titles:
            bigrams_titles[b] = 1
        else:
            bigrams_titles[b] += 1
    for token in tokens:
        if token not in titles_token:
            titles_token[token] = 1
        else:
            titles_token[token] += 1
df_titles_token = pd.DataFrame.from_dict(titles_token,orient="index",columns=["频率"]).sort_values(by="频率",ascending=False)
df_titles_token = df_titles_token[df_titles_token["频率"]>=5]
df_bigrams_titles = pd.DataFrame.from_dict(bigrams_titles,orient="index",columns=["频率"]).sort_values(by="频率",ascending=False)
df_bigrams_titles = df_bigrams_titles[df_bigrams_titles["频率"]>=2]
```
